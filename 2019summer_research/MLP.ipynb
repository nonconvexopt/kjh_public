```python
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#random data generation
rd.seed(0)
class dataset:
    def __init__(self, n):
        self.size = n
        self.x1 = rd.normal(0,1,self.size)
        self.x2 = rd.normal(0,1,self.size)
        self.y = 0.5 - 0.7 * self.x1**2 + 0.1 * self.x2**3 + 0.2 * self.x1*self.x2 + self.x2
    
    def __len__(self):
        return self.size
data = dataset(10000)
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(data.x1, data.x2, data.y)
plt.show()
```
