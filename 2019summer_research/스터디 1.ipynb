```python
"패키지 불러오기"
import tensorflow as tf
import numpy as np
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import matplotlib
import matplotlib.pyplot as plt
```

```python
"예제용 데이터 불러오기"
from sklearn.datasets import load_iris
iris = load_iris()
data = pd.DataFrame(iris['data'],columns=iris['feature_names'])
```

```python
"R처럼 객체 이름 치면 내용 볼 수 있음"
data
```

```python
y = list(data['sepal length (cm)'])
x1 = list(data['sepal width (cm)'])
x2 = list(data['petal length (cm)'])
x3 = list(data['petal width (cm)'])
```

```python
x1
```

```python
"변수 선언"
b0 = tf.Variable(tf.random_uniform([1], -10, 10, dtype=tf.float64, seed=0))
b1 = tf.Variable(tf.random_uniform([1], -10, 10, dtype=tf.float64, seed=0))
b2 = tf.Variable(tf.random_uniform([1], -10, 10, dtype=tf.float64, seed=0))
b3 = tf.Variable(tf.random_uniform([1], -10, 10, dtype=tf.float64, seed=0))
```

```python
tf.Variable(tf.random_uniform([1], -10, 10, dtype=tf.float64, seed=0))
```

```python
"가설 만들기(모델식 만들기)"
y_hat = b0 + b1*x1 + b2*x2 + b3*x3
```

```python
"손실함수 지정"
rmse = tf.sqrt(tf.reduce_mean(tf.square(y-y_hat)))

"최적화 알고리즘 지정 및 하이퍼파라미터 지정"
learning_rate = 0.1
gradient_descent = tf.train.GradientDescentOptimizer(learning_rate).minimize(rmse)
```

```python
"세션 시작 및 학습"
sess = tf.Session()
sess.run(tf.global_variables_initializer())
res_rmse = []
```

```python
for step in range(1, 60):
    sess.run(gradient_descent)
    print("Epoch(step) : %2.f, RMSE = %2.06f, b0 = %.4f, b1 = %.4f, b2 = %.4f, b3 = %.4f"
          % (step,sess.run(rmse),sess.run(b0),sess.run(b1),sess.run(b2),sess.run(b3)))
    res_rmse.append(sess.run(rmse))
```

```python
fig, ax = plt.subplots()
ax.plot(list(range(1,60)),res_rmse)
ax.set(xlabel='epoch(step)', ylabel='rmse')
ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(list(range(17,25)),res_rmse[17:25])
ax.set(xlabel='epoch(step)', ylabel='rmse')
ax.grid()
```
